```python
import requests
from bs4 import BeautifulSoup

url ="https://informatica.ucm.es/profesores-y-tutorias"
req = requests.get(url)

soup = BeautifulSoup(req.text,"html.parser")
soup.title
```

```python
soup.find_all("a") ## hiperenlaces poniendo "a"
for link in soup.find_all("a"):
    print(link.get('href')) ## para limpiar el listado, y solo aparecer la dirección a la que apunta
```

```python
#print(soup.body) ## el codigo html en las webs tienen dos partes: la superior y el cuerpo/body: la primera no se ve en la web,
## y es info para el navegador para indicar cómo tiene que mostrar la info. El body es lo que se muestra.

print(soup.body.prettify()) ## para hacerlo más bonico
```

```python
url = soup.find('iframe').get('src')
url
```

```python
req= requests.get(url)
soup = BeautifulSoup(req.text,"html.parser")
soup
```

```python
headers = soup.find_all('th')

headers[0].text

for header in headers:
    print(header.text)
```

```python
rows = soup.find_all('tr')

data = ['nombre']
for row in rows:
    
    cells = row.find_all('td')
    
    if len(cells)>2:  ## para borrar las listas vacías
         data.append((cells[0].text,
                      cells[1].text,
                      cells[2].text,
                      cells[3].text,
                      cells[4].text))
            
data
```

```python
import pandas as pd

df = pd.DataFrame(data,columns=(headers[0].text.title(),
                                headers[1].text,
                               headers[2].text,
                               headers[3].text,
                               headers[4].text[:3]))

df
```
